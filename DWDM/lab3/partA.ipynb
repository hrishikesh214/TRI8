### DWDM LAB 3 Apriori Algorithm

```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('./market_cart.csv',header=None)
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])

transactions[0]
```

```python
rules = apriori(transactions,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
```

```python
results = list(rules)
df_results = pd.DataFrame(results)
df_results.head()
```

```python
support = df_results.support
```

```python
rows = []

for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    rows.append([",".join(list(single_list[0])),",".join(list(single_list[1])), support[i],single_list[2], single_list[3]])
```

```python
df_final = pd.DataFrame(rows,columns=['LHS','RHS','support','confidence','lift'])
df_final.tail()
```
