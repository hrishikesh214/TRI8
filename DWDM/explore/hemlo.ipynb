```python
import pandas as pd

data = [
    ['Virat', 31, 'Best', 'Batsman', 871, 'IND'], 
    ['Rohit', 33, 'Better', 'Batsman', 855, 'IND'], 
    ['Babar', 25, 'Good', 'Batsman', 829, 'PAK'],
    ['Boult', 31, 'Best', 'Bowler', 722, 'NZ'],
    ['Bumrah', 26, 'Better', 'Bowler', 791, 'IND'],
    ['Mujeeb', 19, 'Good', 'Bowler', 701, 'AFG'],
    ['Nabi', 35, 'Best', 'All Rounder', 301, 'AFG'],
    ['Strocks', 29, 'Better', 'All Rounder', 285, 'ENG'],
]
df = pd.DataFrame(data, columns=['Name', 'Age', 'Grade', 'Role', 'Rating', 'Country'])
```

```python
df.head(n=3)
```

```python
cut_points = [20, 30]
min_age = df['Age'].min()
max_age = df['Age'].max()
cut_labels = ['Junior', 'Experienced', 'Advanced']
df['Age_Group'] = pd.cut(df['Age'], bins=[min_age]+cut_points+[max_age], labels=cut_labels, include_lowest=True)
df
```
