```python
import pandas as pd
```

```python
db = [
    ['A1', 1,1],
    ['A2', -2,1],
    ['A3', -2,-3],
    ['A4', 3,2],
    ['A5', -4,3],
    ['A6', 5,-5],
]
```

```python
# read data
idf = pd.read_csv('../datasets/income.csv')
idb = []
for row in idf.values:
    idb.append(row)
# idb = idb[2:]
db = idf.values.tolist()
```

```python
def get_euclidean_distance(x,y):
    return ((x[0]-y[0])**2 + (x[1]-y[1])**2)**0.5
```

```python
def make_cluster(db, n=2):
    n = [i for i in range(n)] # k1, k2, ...
    clusters = [[] for i in n]
    old_clusters = [[] for i in n]
    distance_table = [[0 for x in n] for i in db]
    centroids = [[db[i][1], db[i][2]] for i in n]
    row_cluster = [0 for i in db]

    while True:
        old_clusters = clusters
        clusters = [[] for i in n]
        for row_index, row in enumerate(db):
            min_distance_cluster = [0, 9999999] # [cluster_index, distance]
            label, x1, y1 = row
            for centroid_index, centroid in enumerate(centroids):
                x2,y2 = centroid
                dis = get_euclidean_distance((x1,y1), (x2,y2))
                distance_table[row_index][centroid_index] = dis
                if dis < min_distance_cluster[1]:
                    min_distance_cluster = [centroid_index, dis]
            clusters[min_distance_cluster[0]].append(row_index)
            row_cluster[row_index] = min_distance_cluster[0]
        
        for __c1, __c2 in enumerate(clusters):
            clusters[__c1].sort()

        if clusters == old_clusters:
            break

        # calculate new centroid

        for __c_index, __c in enumerate(clusters):
            new_x = sum([db[i][1] for i in __c]) / len(__c)
            new_y = sum([db[i][2] for i in __c]) / len(__c)
            centroids[__c_index] = [new_x, new_y]

    final_db = []
    for row_index, row in enumerate(db):
        final_row = row
        distance_table[row_index] = ["{:.2f}".format(i) for i in distance_table[row_index]]
        final_row.append(f'K{row_cluster[row_index]+1}')
        final_row.append(" ,".join(distance_table[row_index]))
        # db[row_index] = row
        final_db.append(final_row)

    return final_db
```

```python
clusters = pd.DataFrame(make_cluster(db, n=3), columns=['Label', 'X', 'Y', 'Cluster', 'Distances'])
clusters
```
